```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.show()
```

```python
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
plt.title('Coding CLub')
plt.show()
```

```python
x1 = [1,2,3]
y1=[2,4,1]


x2 = [1,2,3]
y2 = [4,1,3]

plt.plot(x1,y1, label = "Line-1")
plt.plot(x2,y2, label = "Line-2")
plt.legend()
plt.show()
```

```python
plt.style.use('grayscale')

x = [2,2,3,4,5,6]
y = [2,4,3,5,4,6]

plt.plot(x,y, color='green',linestyle='dashed', linewidth=3, 
         marker = 'o', markerfacecolor='blue', markersize=8)
```

```python
plt.ylim(1,8)
plt.xlim(1,8)
```

## Bar Graph 

```python
#x-cordinates of left sides of bars
left = [1,2,3,4,5,6]

height = [10,24,5,11,36,42]

tick_label = ['one','two','three','four','five','six']

plt.bar(left,height,tick_label = tick_label, width=0.8, color = ['blue', 'red'])
```

## Histogram 

```python
ages = [2,5,70,40,30,45,50,45,43,40,44,60,7,13,57]
range = (0,100)
bins = 10

plt.hist(ages, bins, range, color='yellow', histtype='bar', rwidth=0.8)
```

## Pie-Chart

```python
activities = ['Eat', 'Code', 'Sleep','Class', 'Walk']

slices = [3,7,8,6,4]

colors = ['r','m','g','b','p']

plt.pie(slices,labels=activities, colors=colors, startangle=90,shadow=True,explode=(0,0,0.1,0,0), radius=1.2, autopct = '%l.lf%%')
```
