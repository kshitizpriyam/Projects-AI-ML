```python
#Step-1: To find k-nearest neighbours based on distance
```

```python
#Distance formula based on 2 numpy arrays - Eucidean Distance
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
#KNN Algo
def knn(x,y,querypoint,k=5):
    
    #Pick k-nearest neighbours
    vals = []
    
    #for every point in the x
    for i in range(x.shape[0]):
        #Compute Distance
        d = distance(querypoint,x[i])
        vals.append((d,y[i]))
        
    #print(vals)
    
    #2. Sort the arrays & give k-nearest points
    vals = sorted(vals)
    
    vals = vals[:k]
    
    #Majority Vote
    vals = np.array(vals)
    new_values = np.unique(vals[:,1], return_counts=True)
    #print(new_values)
    
    #Index of MAximum COunt
    index=new_values[1].argmax()
    
    #Map this index with my data
    pred = new_values[0][index]
    
    return pred
```
